###  ‚úÖConvert data frames and drop unneed rows and columns
üü° Could clean up data types and nulls more but the datasets are function as of 3/26/2025

```python
col_order = ['Year','District',  'SchoolName', 'Female', 'Male', 'Hispanic', 'American Indian/Alaskan Native', 'Asian',
       'Black  or African American', 'Caucasian', 'Not Hispanic',
       'Economically Disadvantaged', 'Students with Disabilities',
       'English Language Learners', 'Experiencing Housing Insecurity',
       'Housed']

dist_col_order = ['Year',' District Name', ' School Name', 'Grade Level', 'Enrollment',
       'No. of Student in Tier 1', 'No. of Student in Tier 2',
       'No. of Student in Tier 3', 'No. of Student in Tier 4',
       'Chronically Absent Student', 'Chronic Absentee Rate']
```

```python
from os import rename
for k, v in df_dict.items():
  ##Drop %s and reorder columnson subpop, add district, school
  if k.endswith('subpop'):
    df_dict[k] = df_dict[k].transpose().drop(['Percent of Student in Tier 1','Percent of Student in Tier 2','Percent of Student in Tier 3','Percent of Student in Tier 4','Chronic Absentee Rate'], axis = 0)
    df_dict[k].columns = df_dict[k].iloc[0]
    df_dict[k].drop('Subpopulation Item', inplace = True)

    df_dict[k]['SchoolName'] = k.split('_')[1].split('2')[0]
    df_dict[k]['SchoolName'].replace(school_rename)
    df_dict[k]['District'] = 1
 ## add year column
  if k.split('_')[1].split('2')[1] == '3':
    df_dict[k]['Year'] = 2023
  elif k.split('_')[1].split('2')[1] == '4':
    df_dict[k]['Year'] = 2024
  ## Standardize column order
  if k.endswith('subpop'):
    df_dict[k] = df_dict[k][col_order]
  elif k.endswith('grade'):
    df_dict[k] = df_dict[k][dist_col_order]
```
